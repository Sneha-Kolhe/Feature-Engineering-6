```python
To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you can use the following formula:

�
=
�
(
∑
�
�
)
−
(
∑
�
)
(
∑
�
)
[
�
∑
�
2
−
(
∑
�
)
2
]
[
�
∑
�
2
−
(
∑
�
)
2
]
r= 
[n∑x 
2
 −(∑x) 
2
 ][n∑y 
2
 −(∑y) 
2
 ]
​
 
n(∑xy)−(∑x)(∑y)
​
 

Where:

�
n is the number of observations.
�
x is the amount of time spent studying for the exam.
�
y is the final exam score.
∑
∑ denotes the sum of the values.
​
  denotes the square root.
The Pearson correlation coefficient 
�
r ranges between -1 and 1, where:

�
=
1
r=1 indicates a perfect positive linear relationship.
�
=
−
1
r=−1 indicates a perfect negative linear relationship.
�
=
0
r=0 indicates no linear relationship between the variables.
Now, let's say we have the following data:

Time spent studying (hours)
:
[
3
,
6
,
7
,
4
,
5
]
Final exam scores
:
[
75
,
90
,
85
,
80
,
88
]
Time spent studying (hours)
Final exam scores
​
  
:[3,6,7,4,5]
:[75,90,85,80,88]
​
 
Using these data, we can calculate the Pearson correlation coefficient:

import numpy as np

# Given data
time_spent_studying = [3, 6, 7, 4, 5]
final_exam_scores = [75, 90, 85, 80, 88]

# Calculate the Pearson correlation coefficient
correlation_coefficient = np.corrcoef(time_spent_studying, final_exam_scores)[0, 1]

print("Pearson correlation coefficient:", correlation_coefficient)
Interpreting the result:

If the Pearson correlation coefficient is close to 1, it indicates a strong positive linear relationship between the amount of time spent studying and the final exam scores. This means that as the amount of time spent studying increases, the final exam scores tend to increase.
If the Pearson correlation coefficient is close to -1, it indicates a strong negative linear relationship between the variables. This would imply that as the amount of time spent studying increases, the final exam scores tend to decrease.
If the Pearson correlation coefficient is close to 0, it indicates no linear relationship between the variables, implying that the amount of time spent studying does not significantly affect the final exam scores.
```
